```python
import os
import json
import numpy as np
import pandas as pd
import geopandas as gpd
from geopandas import GeoSeries, GeoDataFrame
import folium as folium
from folium import IFrame
import branca
```

```python
maindata = pd.read_csv('data/masterdata-kuala_lampur.csv')
maindata
```

```python
# All data collected from Zomato Api 

# 100 Best Restaurants in Kuala-Lumpur by Zomato Ratings
df = pd.DataFrame(maindata)

m = folium.Map(location=[3.1390, 101.6869], zoom_start=12)



for i in range(0,len(df)):
    
#     details = 'Name :' + df.iloc[i]['restaurants__restaurant__name'] 
    html = """
    <h4><center>Details</center></h4>
    <p><b>Name :</b> {name}</p>
    <p><b>Cousines :</b> {cousines}</p>
    <p><b>Ave. cost for Two :</b> MYR. {avgcost}</p>
    <p><b>Ratings :</b> {ratings} (Out of 5)</p>
    <p><b>Votes :</b> {votes}</p>
    <p><b>Address :</b> {address}</p>
    """.format(name=df.iloc[i]['restaurants__restaurant__name'],cousines=df.iloc[i]['restaurants__restaurant__cuisines'], \
              avgcost=str(df.iloc[i]['restaurants__restaurant__average_cost_for_two']),ratings=str(df.iloc[i]['restaurants__restaurant__user_rating__aggregate_rating']), \
              votes=str(df.iloc[i]['restaurants__restaurant__user_rating__votes']),address=str(df.iloc[i]['restaurants__restaurant__location__address']) )
    details = branca.element.IFrame(html=html, width=300, height=200)
    
    p = folium.Popup(details, parse_html=True)  
    
    folium.Marker([df.iloc[i]['restaurants__restaurant__location__latitude'], df.iloc[i]['restaurants__restaurant__location__longitude']], popup=p).add_to(m)

m.save('100-Best-Restaurants-in-Kuala-Lumpur-by-Zomato-Ratings.html')
m

# Click on Marker to see details
```
